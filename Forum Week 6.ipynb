```python
#Exercise 1
inventory = {
 'gold' : 500,
 'pouch' : ['flint', 'twine', 'gemstone'],
 'backpack' : ['xylophone','dagger', 'bedroll','bread loaf']
}
#Add a key to inventory called 'pocket'

inventory.setdefault("pocket")

#Set the value of 'pocket' to be a list consisting of the strings 'seashell', 'strange berry', and 'lint'

inventory["pocket"] = ["seashell", "strange berry", "lint"]

#.sort()the items in the list stored under the 'backpack' key.

x = sorted(inventory["backpack"])
inventory["backpack"] = x

#Then .remove('dagger') from the list of items stored under the 'backpack' key.
g = inventory["backpack"]
new = []
for f in g:
 if f != "dagger":
  new.append(f)
inventory["backpack"] = new

#Add 50 to the number stored under the 'gold' key.
h = inventory["gold"]
h = h + 50
inventory["gold"] = h

print(inventory)
```

```python
#Exercise 3
stock = {
 "banana": 6,
 "apple": 0,
 "orange": 32,
 "pear": 15
}
prices = {
 "banana": 4,
 "apple": 2,
 "orange": 1.5,
 "pear": 3
}

def compute_bill(x):
 total = 0
 for t in x:
  y = stock[t]
  g = prices[t]
  if y != 0:
   y = y - 1
   print("Stock :", t, y)
   stock[t] = y
   total = total + g
  else:
   print("Out of Stock for:", t)
 return total

x = 1
bruh =[]
print("type no to Exit")
while x == 1:
 j = input("Enter List of Fruits :")
 bruh.append(j)
 if j == "no":
  x = x -1
  bruh.remove("no")

print("Total : ",compute_bill(bruh))
```

```python
#Exercise 4
eren = {
 "name": "Eren",
 "homework": [90.0,97.0,75.0,92.0],
 "quizzes": [88.0,40.0,94.0],
 "tests": [75.0,90.0]
}
mikasa = {
"name": "Mikasa",
"homework": [100.0, 92.0, 98.0, 100.0],
"quizzes": [82.0, 83.0, 91.0],
"tests": [89.0, 97.0]
}
armin = {
"name": "Armin",
"homework": [0.0, 87.0, 75.0, 22.0],
"quizzes": [0.0, 75.0, 78.0],
"tests": [100.0, 100.0]
}

names = [eren, mikasa, armin]
bruh = ["name", "homework", "quizzes", "tests"]
for x in names:
    for y in bruh:
        print(x[y])

def averege(l):
    total = float(sum(l))
    result = total / len(l)
    return result

def get_averege(dict):
    dict2 = []
    for x in dict.values():
        if type(x) == list:
            dict2.append(x)
    homework = averege(dict2[0])
    quizzes = averege(dict2[1])
    tests = averege(dict2[2])
    total_average = (homework * 10 / 100) + (quizzes * 30 / 100) + (tests * 60 / 100)
    return total_average

def get_letter_grade(x):
    if type(x) == float or type(x) == int:
        if x > 89:
            print("Grade", x,": A")
        elif x > 79:
            print("Grade", x, ": B")
        elif x > 69:
            print("Grade", x, ": C")
        elif x > 59:
            print("Grade", x, ": D")
        else:
            print("Grade", x, ": F")
    if type(x) == list:
        for y in x:
            if y > 89:
                print("Grade", y, ": A")
            elif y > 79:
                print("Grade", y, ": B")
            elif y > 69:
                print("Grade", y, ": C")
            elif y > 59:
                print("Grade", y, ": D")
            else:
                print("Grade", y, ": F")

def get_class_average(b):
    result = []
    for x in b:
        result.append(get_averege(x))
    return result

get_letter_grade(get_class_average(names))
```
